```python
def InsertionSort(arr):# function for insertion sort
 
    
    for i in range(1, len(arr)):# Traverse through 1 to len(arr)
 
        key = arr[i] # variable set for index array element
 
        # Move elements of arr[0..i-1], that are
        # greater than key, to one position ahead
        # of their current position
        j = i-1
        while j >= 0 and key < arr[j] :# finding the insertion point with the sorted array
                arr[j + 1] = arr[j]
                j -= 1
        arr[j + 1] = key
 
 
# Driver code to test above
arr = [12, 11, 13, 5, 6] # example of an unsorted array
InsertionSort(arr) # calling the function
for i in range(len(arr)):
    print ("% d" % arr[i])
    
import pandas as pd
import time 
import statistics 
import numpy as np
from numpy.random import randint 
import matplotlib.pyplot as plt

plt.style.use('ggplot')

# creating arrays for input sizes and execution times : 

Inputsizes = list([100, 250, 500, 750, 1000,1250, 2500, 5000, 10000 ,50000]) # input sizes (n) array
Executiontimes = list([]) # Executiontimes
Num_runs = 10

for i in Inputsizes: # loop to iterate over input size (n) list (lengths)
  
    # Generate random integers 
    array = randint(0, 100, i) # First parameter: Lowest and highest integer range (0 to 99); Second parmeter: Input array size 

    # Algorithm timer:
    start = time.time()     # Start timer
    InsertionSort(array)    # Begin algorithm
    end = time.time()       # Stop timer
    Executiontimes.append(end-start) # Append time

#-------------------------------------------------Algorithm plot----------------------------------------------------------------#

plt.xlabel('Input Size n') # Label x axis
plt.ylabel('Running Time (Milliseconds)')  # Label y axis
plt.plot(Inputsizes,Executiontimes,  '--bo', label ='Insertion Sort') # Plot line points and markers
plt.grid(True) # Add a reference grid 
plt.legend() 
plt.show() 
```
